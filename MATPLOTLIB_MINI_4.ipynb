```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading dataset

```python
df = pd.read_csv('/content/olympic.csv')
```

Top 10 countries with most gold medals

> Add blockquote



```python
df = df[df['Medal'].notna()]
gold_medals = df[df['Medal'] == 'Gold'].groupby('NOC')['Medal'].count().nlargest(10)
plt.figure(figsize=(10, 6))
gold_medals.sort_values().plot(kind='barh', color='gold')
plt.xlabel("Number of Gold Medals")
plt.ylabel("Country")
plt.title("Top 10 Countries with Most Gold Medals")
plt.show()
```

Stacked bar chart for selected countries (USA, China, Russia, UK, Germany)

```python
selected_countries = ['USA', 'CHN', 'RUS', 'GBR', 'GER']
medal_counts = df[df['NOC'].isin(selected_countries)].groupby(['NOC', 'Medal']).size().unstack()
medal_counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=['gold', 'silver', 'brown'])
plt.xlabel("Country")
plt.ylabel("Number of Medals")
plt.title("Medal Comparison of Selected Countries")
plt.legend(title="Medal Type")
plt.show()
```

Total medals won by a country over the years (Example: USA)

```python
country = 'USA'
country_medals = df[df['NOC'] == country].groupby('Year')['Medal'].count()
plt.figure(figsize=(12, 6))
plt.plot(country_medals.index, country_medals.values, marker='o', linestyle='-', color='blue')
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.title(f"Total Medals Won by {country} Over the Years")
plt.grid()
plt.show()
```

Pie chart of medals by sport

```python
sport_medals = df['Sport'].value_counts().nlargest(10)  # Top 10 sports by medal count
plt.figure(figsize=(8, 8))
plt.pie(sport_medals.values, labels=sport_medals.index, autopct='%1.1f%%', startangle=140)
plt.title("Percentage of Medals Won in Different Sports")
plt.show()
```
