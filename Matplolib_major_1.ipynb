```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/content/MM -1 .csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')
```

Convert the 'Order Date' column to datetime format

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

Categorize the data based on date and category

```python
sales_trends = df.groupby(['Order Date', 'Category'])['Sales'].sum().reset_index()
```

Making category columns

```python
sales_pivot = sales_trends.pivot(index='Order Date', columns='Category', values='Sales')
```

Plotting

```python
plt.figure(figsize=(12, 6))
for category in sales_pivot.columns:
  plt.plot(sales_pivot.index, sales_pivot[category], label=category)

plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.title('Sales Trends Over Time by Category')
plt.legend()
plt.grid()
plt.show()
```
