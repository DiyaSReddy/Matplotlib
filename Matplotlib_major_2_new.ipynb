```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the datset

```python
file_path = '/content/covid_19_clean_complete.csv'
df = pd.read_csv(file_path)
```

1. Filter data for specific country

```python
country = "India"
df['Date'] = pd.to_datetime(df['Date'])
df_country = df[df['Country/Region'] == country].groupby('Date')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

1. Line plot for total confirmed cases

```python
plt.figure(figsize=(8, 4))
plt.plot(df_country['Date'], df_country['Confirmed'], label='Confirmed Cases', color='blue')
plt.title('Confirmed Cases')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.legend()
plt.grid()
plt.show()
```

2. Bar plot for total deaths

```python
plt.figure(figsize=(8, 4))
plt.plot(df_country['Date'], df_country['Deaths'], label='Deaths', color='blue')
plt.title('Death Cases')
plt.xlabel('Date')
plt.ylabel('Deaths')
plt.legend()
plt.grid()
plt.show()
```

3. Secondary y-axis for total recoveries

```python
plt.figure(figsize=(8, 4))
plt.plot(df_country['Date'], df_country['Recovered'], label='Recoveries', color='blue')
plt.title('Recoveries')
plt.xlabel('Date')
plt.ylabel('Recoveries')
plt.legend()
plt.grid()
plt.show()
```
