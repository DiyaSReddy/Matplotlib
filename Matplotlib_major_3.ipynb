```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/MM-3.csv")
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df["Month"], df["AvgTemperature"], color="pink", alpha=0.5, edgecolors="violet")
plt.xlabel("month")
plt.ylabel("temp")
plt.title("Temperature variation throughout the yeaar")
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```
