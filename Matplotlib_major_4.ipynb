```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/MM-4.csv")
```

```python
study = 'study score'
exam = ['math score', 'reading score', 'writing score']
plt.figure(figsize=(10, 6))
colors = ['blue', 'red', 'green']
for i, score in enumerate(exam):
    plt.scatter(df[study], df[score], color=colors[i], alpha=0.5, label=score)
plt.xlabel("Study Time (Hours)")
plt.ylabel("Exam Scores")
plt.title("Correlation Between Study Time and Exam Scores")
plt.legend()
plt.grid()
plt.show()
```
