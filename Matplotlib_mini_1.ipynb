```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Superstore.csv')
```

1. Visulaize monthly revenue trends.
Cant be done as date column or details not present in dataset

2. Bar chart for top selling product categories

```python
top_sell = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
top_sell.plot(kind='bar', color='orange')
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Top-Selling Product Categories")
plt.xticks(rotation=45)
plt.show()
```

3: Line Graph for Sales Growth Over Time. Cant be done as date column or details not present in dataset

4: Pie Chart for Percentage of Total Sales by Region

```python
region_sales = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8, 8))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140, colors=['blue', 'orange', 'green', 'red'])
plt.title("Percentage of Total Sales by Region")
plt.show()
```
