```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the dataset

```python
file_path = '/content/GlobalLandTemperaturesByMajorCity.csv'
df = pd.read_csv(file_path)
```

```python
df['dt'] = pd.to_datetime(df['dt'])
df['Year'] = df['dt'].dt.year
year_wise = df.groupby('Year')['AverageTemperature'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(year_wise['Year'], year_wise['AverageTemperature'], marker='o', linestyle='-', color='b')
plt.title('Global Average Temperature Change Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (°C)')
plt.grid(True)
plt.show()
```
